```python
# Importing PyCaret's regression module
from pycaret.regression import *

# Setup the environment with advanced options
regression_setup = setup(
    data=train,  # Replace 'train' with your actual dataset
    target='AQI Value',  # Specify your target column
    session_id=123,  # Seed for reproducibility
    normalize=True,  # Normalize numeric data
    polynomial_features=True,  # Add polynomial features to capture non-linear relationships
    remove_multicollinearity=True,  # Drop highly correlated features
    multicollinearity_threshold=0.95,  # Threshold for correlation filtering
)

# Compare models and select the best one
best_model = compare_models(sort='MAE')  # Sort by Mean Absolute Error (MAE)

```

```python
tuned_et_model = tune_model(best_model, optimize='MAE')
```

```python
ensemble_et_model = ensemble_model(tuned_et_model, method='Bagging')
```

```python
# Finalize the original Extra Trees Regressor model
final_model = finalize_model(best_model)

```
